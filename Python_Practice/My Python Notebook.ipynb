```python
def end_other(a, b):
    if a.lower()[-1] == b.lower()[-1]:
      min1 = a if len(a)  < len(b) else b
      max1 = a if len(a)  > len(b) else b
      if min1 == max1[len(max1)-len(min1):]:
        return True
    return False
 
```

```python
def sum13(nums):
    if len(nums) == 0:
        return 0
    elif '13' not in nums:
        return sum(nums)
    else:
        for i in nums:
            return sum(nums[:i])
```

```python
sum13([1, 2, 13, 2, 1, 13])
```

```python
def sum67(nums):
    for i in range(0,len(nums)):
        if nums[i] == 6:
            nums[i] = 0
            for j in range(i+1, len(nums)):
                if nums[j] == 7:
                    nums[i+1:j] = 0
        return sum(nums)
```
